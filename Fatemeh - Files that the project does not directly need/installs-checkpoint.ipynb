```python
from duckduckgo_search import DDGS
from fastcore.all import L

def search_images_ddg(term, max_images=50):
    """Search DuckDuckGo for images and return a list of image URLs."""
    print(f"Searching for '{term}' ...")
    urls = []
    with DDGS() as ddgs:
        for r in ddgs.images(keywords=term, max_results=max_images):
            urls.append(r["image"])
    return L(urls)
```

```python
from pathlib import Path
from PIL import Image
import shutil

def unify_extensions_to_jpg_inplace(
    folder: Path,
    make_backup: bool = True,      # بکاپ بسازد؟
    delete_png_after: bool = True, # بعد از تبدیل png حذف شود؟
    dry_run: bool = True,
    quality: int = 95,             # کیفیت jpg برای تبدیل png -> jpg
):
    folder = Path(folder)
    assert folder.exists(), f"Folder not found: {folder}"

    backup_dir = folder / "_originals_backup"
    if make_backup and not dry_run:
        backup_dir.mkdir(exist_ok=True)

    counts = {
        "renamed_jpeg_to_jpg": 0,
        "converted_png_to_jpg": 0,
        "moved_to_backup": 0,
        "skipped_exists": 0,
        "errors": 0,
    }

    for f in folder.iterdir():
        if not f.is_file():
            continue
        if f.name == ".gitkeep":
            continue
        if f.parent.name == "_originals_backup":
            continue

        suf = f.suffix.lower()

        # --- 1) .jpeg or typo .jepg  -> rename to .jpg (safe, no re-encode)
        if suf in {".jpeg", ".jepg"}:
            out = f.with_suffix(".jpg")

            # اگر out وجود دارد، این فایل را ببریم بکاپ
            if out.exists():
                if make_backup:
                    target = backup_dir / f.name
                    if dry_run:
                        print(f"[PLAN] MOVE (exists): {f.name} -> _originals_backup/")
                    else:
                        # جلوگیری از تداخل در بکاپ
                        if target.exists():
                            target = backup_dir / f"{f.stem}__dup{f.suffix}"
                        shutil.move(str(f), str(target))
                    counts["moved_to_backup"] += 1
                else:
                    if dry_run:
                        print(f"[PLAN] SKIP (exists): {f.name}  (because {out.name} exists)")
                    counts["skipped_exists"] += 1
                continue

            if dry_run:
                print(f"[PLAN] RENAME: {f.name} -> {out.name}")
            else:
                f.rename(out)
            counts["renamed_jpeg_to_jpg"] += 1
            continue

        # --- 2) .png -> convert to REAL jpg (needs re-encode)
        if suf == ".png":
            out = f.with_suffix(".jpg")

            if out.exists():
                # اگر jpg هست، png اضافه است → بکاپ یا حذف
                if make_backup:
                    target = backup_dir / f.name
                    if dry_run:
                        print(f"[PLAN] MOVE PNG (jpg exists): {f.name} -> _originals_backup/")
                    else:
                        if target.exists():
                            target = backup_dir / f"{f.stem}__dup{f.suffix}"
                        shutil.move(str(f), str(target))
                    counts["moved_to_backup"] += 1
                else:
                    if dry_run:
                        print(f"[PLAN] SKIP PNG (jpg exists): {f.name}")
                    counts["skipped_exists"] += 1
                continue

            if dry_run:
                print(f"[PLAN] CONVERT: {f.name} -> {out.name}")
            else:
                try:
                    img = Image.open(f)
                    # اگر alpha داشت، سفیدش می‌کنیم
                    if img.mode in ("RGBA", "LA"):
                        bg = Image.new("RGB", img.size, (255, 255, 255))
                        bg.paste(img, mask=img.split()[-1])
                        img = bg
                    else:
                        img = img.convert("RGB")

                    img.save(out, format="JPEG", quality=quality, optimize=True)

                    if delete_png_after:
                        f.unlink(missing_ok=True)

                except Exception as e:
                    print("[ERROR] convert png:", f.name, e)
                    counts["errors"] += 1
                    continue

            counts["converted_png_to_jpg"] += 1
            continue

        # بقیه پسوندها کاری نداریم (مثل .jpg موجود)
        # اگر خواستی webp هم اضافه کنیم، می‌گم چطور.

    print("\nSUMMARY:", counts)
    if dry_run:
        print("Dry-run done. Set dry_run=False to execute.")
    return counts


# ===== RUN on your folder =====
folder = Path(r"C:\Users\Elahe\Desktop\Repo-ML-PRO\Repo-xAI-Proj-B\data\robustness\YSQD\coffee-mug")

# 1) Dry run
unify_extensions_to_jpg_inplace(folder, dry_run=True)

# 2) Real run
# unify_extensions_to_jpg_inplace(folder, dry_run=False)
```

```python
from pathlib import Path
import shutil

folder = Path(r"C:\Users\Elahe\Desktop\Repo-ML-PRO\Repo-xAI-Proj-B\data\robustness\YSQD\coffee-mug")
backup_dir = folder / "_originals_backup"

dry_run = True  # اول True، بعد False

# جمع کردن فایل‌ها بدون تکرار (بدون glob دوگانه)
jpeg_files = sorted({
    p for p in folder.iterdir()
    if p.is_file()
    and p.name != ".gitkeep"
    and p.suffix.lower() in {".jpeg", ".jepg"}  # هم jpeg هم تایپی jepg
})

print("Found jpeg/jepg:", len(jpeg_files))

if not dry_run:
    backup_dir.mkdir(exist_ok=True)

for src in jpeg_files:
    # اگر در همین اجرای برنامه قبلاً rename شده باشد، دیگر وجود ندارد → رد کن
    if not src.exists():
        continue

    dst_jpg = src.with_suffix(".jpg")  # 10.jpeg -> 10.jpg

    if dst_jpg.exists():
        # اگر 10.jpg از قبل وجود دارد، jpeg را به بکاپ منتقل کن (تا چیزی از دست نرود)
        dst = backup_dir / src.name
        if dst.exists():
            dst = backup_dir / f"{src.stem}__dup{src.suffix}"

        print("[MOVE]  ", src.name, "->", dst.name)
        if not dry_run:
            shutil.move(str(src), str(dst))
    else:
        print("[RENAME]", src.name, "->", dst_jpg.name)
        if not dry_run:
            src.rename(dst_jpg)

print("\nRecount:")
print("jpg :", len(list(folder.glob("*.jpg"))))
print("jpeg:", len(list(folder.glob("*.jpeg"))))
print("jepg:", len(list(folder.glob("*.jepg"))))
print("png :", len(list(folder.glob("*.png"))))
```
